```python
# Installing Gradio
!pip install streamlit transformers -q
# Import the necessary libraries
import streamlit as st
import numpy as np
import pandas as pd
import pickle
import transformers
from transformers import AutoTokenizer, AutoConfig,AutoModelForSequenceClassification,TFAutoModelForSequenceClassification, pipeline
from scipy.special import softmax

# Define the NLP pipeline
nlp_pipeline = pipeline("sentiment-analysis", model="Achar/NLP-Sentinel-Analysis")

# Define the Streamlit app
def main():
    st.title("NLP Sentiment Analysis App")

    # Add a text input field
    text = st.text_input("Enter some text:")

    if text:
        # Use the NLP pipeline to predict the sentiment of the text
        prediction = nlp_pipeline(text)[0]

        # Display the prediction
        st.write(f"Sentiment: {prediction['label']}")
        st.write(f"Score: {prediction['score']}")

# Run the app
if __name__ == "__main__":
    main()
```
